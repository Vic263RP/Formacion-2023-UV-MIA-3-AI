```python
import pandas as pd
```

```python
s_train = pd.read_csv('./data/simple_example/training.csv', index_col="Datetime",
                      parse_dates=True, squeeze=True)
print(s_train)
```

```python
# %load_ext autoreload
# %autoreload 2
```

```python
from adtk.data import validate_series
s_train = validate_series(s_train)
print(s_train)
```

```python
from adtk.visualization import plot

plot(s_train);
```

```python
from adtk.detector import SeasonalAD

seasonal_ad = SeasonalAD()
anomalies = seasonal_ad.fit_detect(s_train)
plot(s_train, anomaly=anomalies, anomaly_color="red", anomaly_tag="marker");
```

```python
known_anomalies = pd.read_csv("./data/simple_example/known_anomalies.csv", index_col="Datetime", parse_dates=True, squeeze=True)
known_anomalies = validate_series(known_anomalies)
```

```python
from adtk.data import to_events

known_anomalies = to_events(known_anomalies)
print(known_anomalies)
plot(s_train,
     anomaly={"Known": known_anomalies, "Model": anomalies},
     anomaly_tag={"Known": "span", "Model": "marker"},
     anomaly_color={"Known": "orange", "Model": "red"});
```

```python
s_test = pd.read_csv("./data/simple_example/testing.csv", index_col="Datetime",
                     parse_dates=True, squeeze=True)
s_test = validate_series(s_test)

print(s_test)

anomalies_pred = seasonal_ad.detect(s_test)
plot(s_test, anomaly=anomalies_pred, ts_linewidth=1, anomaly_color='red',
     anomaly_tag="marker");
```
