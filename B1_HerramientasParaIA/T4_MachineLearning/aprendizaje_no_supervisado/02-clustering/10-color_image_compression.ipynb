# Vector Quantization Example (2)

A color 1024 x 768 size image of a raccoon is used here to illustrate how `k`-means is used for vector quantization.

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

from sklearn import cluster
```

```python
from scipy.misc import face
face = face(gray=False)

plt.figure(figsize=(5, 5))
plt.imshow(face);
```

```python
print(face.shape)

768*1024
```

```python
n_clusters = 50
np.random.seed(0)

X = face.reshape((-1, 3))  # We need an (n_sample, n_feature) array
X.shape
```

```python
k_means = cluster.KMeans(n_clusters=n_clusters, n_init=4)
k_means.fit(X)
values = k_means.cluster_centers_
labels = k_means.labels_
```

```python
values
```

```python
values.shape
```

```python
face_compressed = face.copy()
ancho = face.shape[0]
alto  = face.shape[1]
for i in range(ancho):
    for j in range(alto):
        face_compressed[i,j] = values[labels[i*alto+j]]
```

```python
# original face
plt.figure()
plt.imshow(face);
```

```python
# compressed face
plt.figure()
plt.imshow(face_compressed);
```
